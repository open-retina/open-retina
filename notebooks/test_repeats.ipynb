```python
import os
import pickle
import matplotlib.pyplot as plt
import torch
import numpy as np
import itertools

import openretina
from openretina.models.core_readout import CoreReadout
from openretina.hoefling_2024.data_io import (
    natmov_dataloaders_v2,
)
from openretina.neuron_data_io import make_final_responses
from openretina.utils.h5_handling import load_h5_into_dict
from openretina.measures import CorrelationLoss3d, PoissonLoss3d, MSE3d
data_folder = os.path.expanduser("/gpfs01/euler/data/SharedFiles/projects/TP12/")
movies_path = os.path.join(data_folder, "2024-01-11_movies_dict_8c18928.pkl")
```

```python
# read movies
with open(movies_path, "rb") as f:
    movies_dict = pickle.load(f)

data_path_responses = os.path.join(data_folder, "2024-08-14_neuron_data_responses_484c12d_djimaging.h5")
responses = load_h5_into_dict(data_path_responses)

data_dict = make_final_responses(responses, response_type="natural")  # type: ignore
dataloaders, neuron_variances = natmov_dataloaders_v2(
    data_dict, movies_dictionary=movies_dict, train_chunk_size=100, seed=1000)
```

```python
corr_loss = CorrelationLoss3d(avg=True)
pois_loss = PoissonLoss3d(avg=False)
mse_loss = MSE3d()
session_id = sorted(dataloaders["test"].keys())[4]
test_responses_by_trial = dataloaders["test"][session_id].dataset.test_responses_by_trial
print(test_responses_by_trial.shape)

colors = ["blue", "red", "orange"]
for neuron_id in range(20,40):
    repeats = list(range(test_responses_by_trial.shape[1]))
    for i in repeats:
        plt.plot(test_responses_by_trial[neuron_id, i], color=colors[i], alpha=0.6)
    plt.plot(test_responses_by_trial[neuron_id].mean(axis=0), color="green")

    rqi = torch.var(torch.mean(test_responses_by_trial[neuron_id], axis=0)) / torch.mean(torch.var(test_responses_by_trial[neuron_id], axis=-1))
    uns_traces = [test_responses_by_trial[neuron_id].mean(axis=0).unsqueeze(0).unsqueeze(-1)] + [test_responses_by_trial[neuron_id, i].unsqueeze(0).unsqueeze(-1) for i in repeats]
    print(f"## {neuron_id=}: {rqi=:.1%}")
        
    for i, k in itertools.combinations(repeats + [len(repeats)], 2):
        corr = -corr_loss(uns_traces[i], uns_traces[k]).item()
        pois = pois_loss(uns_traces[i], uns_traces[k]).item()
        mse = mse_loss(uns_traces[i], uns_traces[k]).item()

        repeat_str = f"repeat{i}-repeat{k}".replace("repeat0", "mean")
        print(f"{repeat_str}: {corr=:.1%} {pois=:.1f}, {mse=:.1f}")
    plt.show()
```
